### Unit converter
Points to observe:  
1.Printing menu to choose operations in a program  
2.Method of adding multiple functionality to a program 

```python
def print_menu():
    print('1.Kilometer to Miles')
    print('2.Miles to Kilometer')
    print('3.Inch to centimeter')
    print('4.Centimeter to Inch')

def km_miles():
    km=float(input('Enter distance in kilometers:'))
    miles = km/1.609
    print('Distance in miles: {0:.3f}'.format(miles))

def miles_km():
    miles=float(input('Enter distance in miles:'))
    km = miles * 1.609
    print('Distance in kilometers: {0:.3f}'.format(km))
    
def inch_cm():
    inch=float(input('Enter the length in inches:'))
    cm=inch*2.54
    print('Length in centimeter: {0:.3f}'.format(cm))

def cm_inch():
    cm=float(input('Enter the length in centimeters:'))
    inch=cm/2.54
    print('Length in inch: {0:.3f}'.format(inch))

print_menu()
try:
    choice=int(input('Enter menu number:'))
    if choice==1:
        km_miles()
    elif choice==2:
        miles_km()
    elif choice==3:
        inch_cm()
    elif choice==4:
        cm_inch()
except ValueError:
    print('please enter a valid menu number')
```

### Quadratic equation solver

Points to observe:  
1.Rerun a program until recieving a halt command

```python
def quad_roots(a,b,c):
    D=(b*b-4*a*c)**(0.5)
    x1=(-b+D)/(2*a)
    x2=(-b-D)/(2*a)
    print('x1: {0:.3f}'.format(x1))
    print('x2: {0:.3f}'.format(x1))

check='y'
while(check=='y'):
    a=float(input('Enter a:'))
    b=float(input('Enter b:'))
    c=float(input('Enter c:'))
    quad_roots(a,b,c)
    check=input('Rerun?(y/n)')
```

### relationship between gravitational force and distance between two bodies

Point to observe:  
1.A seperate function that has all the aesthetic features of the graph  
2.A seperate function for calcuating the values and calling the graphing function

```python
import matplotlib.pyplot as plt

def draw_graph(x,y):
    plt.plot(x,y,marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in Newtons')
    plt.title('Gravitational force and distance')
    plt.grid()
    plt.show()

def generate_F_r(a,b):
    r=range(a,b,50)
    G=6.674*(10**-11)
    m1=0.5
    m2=1.5
    F=[G*(m1*m2)/dist**2 for dist in r]
    draw_graph(r,F)

check='y'
while(check=='y'):
    r1=float(input('Enter r initial:'))
    r2=float(input('Enter r final:'))
    generate_F_r(int(r1),int(r2)) if r1.is_integer() and r2.is_integer() else print('distance are not integers')
    check=input('Rerun?(y/n)')
```

### projectile motion

```python
from matplotlib import pyplot as plt
import math

def xaxis(start,final,increment):
    numbers=[]
    while start<final: numbers.append(start);start=start+increment;
    return numbers

def draw_graph(x,y):
    plt.plot(x,y,marker='o')
    plt.xlabel('land (meters)')
    plt.ylabel('Height (meters)')
    plt.title('Projectile motion of the ball')

    
def trajectory(u,theta):
    theta = math.radians(theta)
    g=9.8
    
    t_flight=2*u*math.sin(theta)/g
    time=xaxis(0,t_flight,0.05)
    x=[]
    y=[]
    x=[u*math.cos(theta)*t for t in time]
    y=[(u*math.sin(theta)*t)-0.5*g*t*t for t in time]
    draw_graph(x,y)

check='y'
while(check=='y'):
    try:
        u=input('Enter the initial velocities (m/s) (comma seperated):')
        theta=input('Enter the angles of projection (degrees):')
        u=list(map(int,u.split(',')))
        theta=list(map(int,theta.split(',')))
    except ValueError:
        print('You have entered an invalid input')
    else:
        plt.figure(figsize=(15,10))
        leg=[]
        for speed in u:
            for angle in theta:
                trajectory(speed,angle)
                leg.append(str(speed)+'m/s'+','+str(angle)+'deg')

    plt.grid()
    plt.legend(leg)
    plt.show()
    check=input('Rerun?(y/n)')
```

### Calculating mean of data

points to observe:  
1.Using sort function to calculate median  
2.Using collections library to determine the list of most occuring values  
3.Finding the dispersion of data  
4.Pearson correlation coefficient measures the strength of the linear relationship. For measuring strength of nonlinear relationships, other method have to be used. So 0 Pearson correlation number implies that there is no linear relationship only, not independence 

```python
from collections import Counter

def calMean(var):
    return sum(var)/len(var)
    
def calMedian(var):
    var.sort()
    median=calMean([var[int(len(var)/2)],var[int((len(var)-1)/2)]]) if len(var)%2==0 else var[int((len(var)-1)/2)]
    return median 

def calMode(var):
    c=Counter(var)
    numFreq=c.most_common()
    maxCount=numFreq[0][1]
    modes=[]
    for numbers in numFreq:
        if numbers[1]==maxCount:
            modes.append(numbers[0])
    return modes

def frequencyTable(var):
    c=Counter(var)
    numFreq=c.most_common()
#     numFreq.sort()
    print('Number\tFrequency')
    for num,freq in numFreq:
        print('{0}\t{1}'.format(num,freq))
        
# measuring dispersion of the data

def findRange(var):
    lowest=min(var)
    highest=max(var)
    dataRange = highest-lowest
    return lowest,highest,dataRange

def variance(var):
    mean=calMean(var)
    diff = [(mean-entry)**2 for entry in var]
    return sum(diff)/len(var)
    

data = [100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200]
print('Mean data over the last {0} days is {1}'.format(len(data),calMean(data)))
print('Median of data over the last {0} days is {1}'.format(len(data),calMedian(data)))
print('Mode of data over the last {0} days is {1}'.format(len(data),calMode(data)))
frequencyTable(data)
print('Lowest: {0} Highest: {1} Range: {2}'.format(findRange(data)[0],findRange(data)[1],findRange(data)[2]))
print('The variance of the data is {0:.4f}'.format(variance(data)))
print('The standard deviation of the data is {0:.4f}'.format(variance(data)**0.5))
```

```python
def correlation(var1,var2):
    if len(var1)==len(var2):
        n=len(var1)
        x=var1
        y=var2
        xy=[var1[i]*var2[i] for i in range(n)]
        x2=[var1[i]*var1[i] for i in range(n)]
        y2=[var2[i]*var2[i] for i in range(n)]
        corr=(n*sum(xy)-sum(x)*sum(y))/(((n*sum(x2)-(sum(x))**2)*(n*sum(y2)-(sum(y))**2))**0.5)
        return corr
    else:
        print('The lengths of data dont match')
        return 0

import pandas as pd
grades=pd.read_csv('grades.csv')
print('Correction between data: {0:.3f}'.format(correlation(grades.iloc[:,0],grades.iloc[:,1])))

plt.scatter(grades.iloc[:,0],grades.iloc[:,1])
plt.title('strong Correlation means nearly a straight line will appear between both the data here')
plt.xlabel(list(grades.columns)[0])
plt.ylabel(list(grades.columns)[1])
plt.grid()
```
