Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import math
# np.linspace?
x=np.linspace(-2*math.pi,2*math.pi,100)
f=np.sin(x)
g=np.cos(x)
h=x**2+2*x

plt.plot(x,f,'og')
plt.plot(x,g,'--b',linewidth=10)
# pl.plot(x,h,'*r')
plt.grid()
plt.title(r"Example title: $x^2-5$") # latex format
plt.xlabel(r"$x$")
plt.ylabel(r"$function$")
plt.annotate("$local minima$",xy=(math.pi,np.cos(math.pi)))
plt.annotate("local maxima",xy=(0,np.cos(0)))
plt.xlim(-4,5)
plt.ylim(-2,2)
# plt.plot?
# plt.ylabel?
```
